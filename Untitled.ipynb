```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
headers = {
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Safari/537.36'
    }
url = 'https://www.mangoplate.com/restaurants/n9ZwosnPSP'
html = requests.get(url, headers=headers).text
soup = BeautifulSoup(html, 'html.parser')
```

```python
name = soup.find(class_='restaurant_name').text.strip().replace('\n','')
print(name)
```

```python
star = soup.find(class_='rate-point').text.strip().replace('\n','')
print(star)
```

```python
abbress = soup.select_one('.info>tbody>tr:nth-child(1)>td').text.strip().replace('\n', ' ')
phone_number = soup.select_one('.info>tbody>tr:nth-child(2)>td').text.strip().replace('\n', ' ')
food = soup.select_one('.info>tbody>tr:nth-child(3)>td').text.strip().replace('\n', ' ')
sale = soup.select_one('.info>tbody>tr:nth-child(4)>td').text.strip().replace('\n', ' ')
parking = soup.select_one('.info>tbody>tr:nth-child(5)>td').text.strip().replace('\n', ' ')
time = soup.select_one('.info>tbody>tr:nth-child(6)>td').text.strip().replace('\n', ' ')

print(abbress)
print(food)
print(phone_number)
print(sale)
print(parking)
print(time)
```
